### Tuple과 list의 차이점
|   	                     | 튜플(tuple)   	| 리스트(list)  	| 
|---	                     |---	|---	|
|길이                  |고정길이   	|가변길이   	|
|아이템 추가   	|불가능 - append/extend 없음   	| 가능 - append/extend 있음  |
|아이템 제거   	|불가능 - remoe/pop 없음   	| 가능 - remoe/pop 있음(del도 가능)  	|
| 속도  	             |빠름 - 이터레이션에 유리   	| 느림(상대적)  	|
|아이템 타입   	|원소로 다양한 타입을 저장하는 경우 주로 사용(struct와 비슷)   	|원소로 단일 타입을 저장하는 경우 주로 사용(array와 비슷)   	|

## 집합(Set)

특징
 * 집합은 중복된 데이터를 가질 수 없다.
 * 순서가 없다. 
 
 
 따라서 순서와 관련된 인덱스기호([ ])를 사용할 수 없고, <br>
 중복 데이터를 만드는 +, *를 사용할 수 없습니다. <br>하지만, in, not in, len()은 사용할 수 있습니다.
 
 리스트나 튜플은 순서가 있기(ordered) 때문에 인덱싱을 통해 자료형의 값을 얻을 수 있지만 set 자료형은 순서가 없기(unordered) 때문에 인덱싱으로 값을 얻을 수 없다. 딕셔너리와 비슷하다. 딕셔너리 역시 순서가 없는 자료형이라 인덱싱을 지원하지 않는다. 만약 set 자료형에 저장된 값을 인덱싱으로 접근하려면 다음과 같이 리스트나 튜플로 변환한 후 해야 한다.

````python
# my_set이라는 빈 집합을 만들고 타입을 출력하세요..
my_set = set()
type(my_set)
````

### 합집합, 교집합, 차집합, 대칭차집합, 부분집합, 포함관계

````python
A = {1, 2, 3, 4, 5, 6}
B = {1, 2, 3, 7, 8, 9}

# 합집합을 구하세요
print(A.union(B))

# 교집합을 구하세요
print(A.intersection(B))

# 차집합을 구하세요
print(A.difference(B))

# 대칭차집합을 구하세요
print(A.symmetric_difference(B))

# 부분집합
# {4, 5, 6}가 A의 부분집합인지 확인하세요
print({4,5,6}.issubset(A))

# A가 {4, 5, 6}을 포함하닌지 확인하세요
print(A.issuperset({4,5,6}))
````

````python
s_float = {1.0, 3.0, 2.0, 10.0}
print(s_float)
print()

# s_float의 길이를 구하세요
len(s_float)

# s_float를 올림차순 정렬하세요
sorted(s_float)

# s_float에 2.0이 포함되는지 확인하세요
{2.0}.issubset(s_float)
````

## 사전(Dictionary)
* 동일한 Key가 2개 존재할 경우 같은 쌍은 무시
* Key에 리스트는 쓸 수 없다는 것이다. 하지만 튜플은 Key로 쓸 수 있다. 

````python
solar1 = ['태양', '수성', '금성', '지구', '화성', '목성', '토성', '천왕성', '해왕성']
solar2 = ['Sun', 'Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
solardict = {}

# 아래와 같이 출력되도록 코딩하세요.
# for문을 사용하세요.
# 
# {'태양': 'Sun', '수성': 'Mercury', '금성': 'Venus', '지구': 'Earth', 
#  '화성': 'Mars', '목성': 'Jupiter', '토성': 'Saturn', '천왕성': 'Uranus', '해왕성': 'Neptune'}

for i,k in enumerate(solar1):
    solardict[k] = solar2[i]
    
print(solardict)
````

#### enumerate
반복문 사용 시 몇 번째 반복문인지 확인이 필요할 수 있습니다. 이때 사용합니다.
인덱스 번호와 컬렉션의 원소를 tuple형태로 반환합니다.

```
>>> t = [1, 5, 7, 33, 39, 52]
>>> for p in enumerate(t):
...     print(p)
... 
(0, 1)
(1, 5)
(2, 7)
(3, 33)
(4, 39)
(5, 52)
```


tuple형태 반환을 이용하여 아래처럼 활용할 수 있습니다.\


```
>>> for i, v in enumerate(t):
...     print("index : {}, value: {}".format(i,v))
... 
index : 0, value: 1
index : 1, value: 5
index : 2, value: 7
index : 3, value: 33
index : 4, value: 39
index : 5, value: 52
```

### dictionary 값 출력 및 삭제

````python
names = {'Mary':10999, 'Sams':2111, 'Aimy':9778, 'Tom':20245, 
'Michale':27115, 'Bob':5887, 'Kelly':7855}

# 'Sams'를 삭제하세요
del names['Sams']
print(names)

# value 값들의 합을 구하세요
n_sum=0
for val in names.values():
    n_sum = n_sum + val
print(n_sum)

# value 값들의 최대값을 구하세요
print(max(names.values()))

# key값들을 출력하세요
print(list(names.keys()))
# value값들을 출력하세요
print(list(names.values()))
# 모든 요소를 삭제하세요
names.clear()
print(names)
````

### dictionary 키 혹은 값 정렬

````python

# 1. key를 기준으로 사전식 오름차순 정렬
names = {'Mary':10999, 'Sams':2111, 'Aimy':9778, 'Tom':20245, 
'Michale':27115, 'Bob':5887, 'Kelly':7855}

print(sorted(names.keys())) # 반환값이 key만 나옴. 영구적인 sort가 아님


# 2. values을 기준으로 오름차순 정렬
names = {'Mary':10999, 'Sams':2111, 'Aimy':9778, 'Tom':20245, 
'Michale':27115, 'Bob':5887, 'Kelly':7855}
ret1 = sorted(names.items(), key=lambda x: x[1])
print(ret1) # 반환값이 사전식으로 나옴
 
# 3. 2번과 비슷
ret2 = sorted(names.values())
print(ret2) # 반환되는값이 value값만 잇음
````

### dictionary 값 설정

````python
my_idol = {"name": "쯔위", "age": 18, "job": "singer"}

# age의 값을 출력하시오
print(my_idol['age'])

# job의 값을 출력하시오
print(my_idol['job'])

# name의 값을 '수지'로 수정하시오
my_idol['name'] = '수지'
print(my_idol)

# 'birth' 의 값을 '1994-07-07'로 추가하시오
my_idol.setdefault('birth','1994-07-07')
print(my_idol)

# 'manager'의 값을 'JYP'로 추가하시오
my_idol.setdefault('manager','JYP')
print(my_idol)
````

````python
my_idol = {'job': 'singer', 'age': 18, 'name': '수지', 'birth': '1994-10-10'}
print(my_idol)
print()

# 'birth' 항목 가져오고 삭제하기
my_idol.pop('birth')
print(my_idol)

# 'age' 항목 삭제하기
del my_idol['age']
print(my_idol)

# 사전 비우기
my_idol.clear()
print(my_idol)

````
