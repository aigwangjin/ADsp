## 텍스트 인덱싱
1. 맨앞이 0부터 시작.<br>
3:6은 3번째부터 6번째까지<br> :6은 0번째부터 6번째까지
2. 맨뒤는 음수일때  카운트<br>
3. 2칸씩 건너뛰어져서 보여줌<br>
4. 음수는 뒤에서 부터 실행되므로 역수로 출력<br>

```python
txt1 = 'A tale that was not right'
txt2 = '이 또한 지나가리라.'
# 0부터 시작
print(txt1[3:6]) # ‘ale ’이 출력됨 
print(txt1[:6])  # ‘A tale’ 이 출력됨
print(txt2[-4:]) # ‘가리라.’가 출력됨
print(txt2[::2])
print(txt2[::-1])
```

## 텍스트 format 설정
format함수<br>
{0} 부분에 원하는 값을 넣을 수 있다.<br>
1개만 있으면 {}만 해도 되고 {0}, {1}, {2} ... 쓸수있다

```python
a=input('저장할 파일 이름을 입력하세요: ')
s='당신이 저장한 파일은 <{0}.jpg> 입니다'
s.format(a)
```

```python
ip = input('임의의 문장을 입력하세요: ')
s = '당신이 입력한 문장에는 {}가 {}습니다.'
if 'a' in ip:
    print(s.format('a','있'))
else:
    print(s.format('a','없'))
```

문장에서 전부 대분자, 소문자로 만들 수 있다.
upper, lower 함수

```python
txt = 'A lot of Things occur each day.'

       # ‘A LOT OF THINGS OCCUR EACH DAY.’가 출력됨
       # ‘a lot of things occur each day.’가 출력됨
print(txt.upper())
print(txt.lower())
```

## 텍스트 공백제거
strip

```python
txt = '   양쪽에 공백이 있는 문자열입니다.   '

print('<'+txt+'>')
ret1 = txt.lstrip()
ret2 = txt.rstrip()
ret3 = txt.strip()
print('<'+ret1+'>')
print('<'+ret2+'>')
print('<'+ret3+'>')
```

## 텍스트 내 특정문자 갯수 출력
count 함수

```python
txt = 'A lot of things occur each day, every day.'

#문자 'o'의 갯수 출력
print(txt.count('o'))
```

## 텍스트내 특정문자의 인덱스값을 출력

## 리스트 사이에 삽입할 문자 선택
join 함수

```python
loglist = ['2016/08/26 10:12:11', '200', 'OK', '이 또한 지나가리라']
bond = ';'
log = bond.join(loglist)
print(log)
```
